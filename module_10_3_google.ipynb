# Домашняя работа "Банковские операции"

```python
# Загрузим необходимые библиотеки
import threading
import time
from random import randint
# Объявим блокировку
lock = threading.Lock()
```

```python
# Создадим класс
class Bank:
  def __init__(self):
    self.balance = 0
  # Процедура пополнения баланса
  def deposid(self):
    for __ in range(100):
      rand_many = randint(50, 500)
      self.balance = self.balance + rand_many
      print(f'Пополнение {rand_many}. Баланс: {self.balance}')
      # Снятие блокировки
      if self.balance >= 500 and lock.locked():
        print(f'Счет разблокирован для снятия. Баланс: {self.balance}')
        lock.release()
      time.sleep(0.001)
  # Процедура снятия
  def take(self):
    for __ in range(100):
      take_many = randint(50, 500)
      print(f'Запрос на {take_many}')
      # Если денег хватит снимаем
      if take_many <= self.balance:
        self.balance = self.balance - take_many
        print(f'Снятие {take_many}. Баланс: {self.balance}')
      # Если не хватает то счет блокируем
      else:
        lock.acquire()
        print('Операция заблокирована не достаточно средств')
      time.sleep(0.001)
```

```python
# Создаем экземпляр банка
bk = Bank()
# Создаем потоки
thread_1 = threading.Thread(target=Bank.deposid, args=(bk, ))
thread_2 = threading.Thread(target=Bank.take, args=(bk, ))
# Стартуем потоки
thread_1.start()
thread_2.start()
```

# Проработка лекции "Проблемы многопоточности и блокировки"

```python
# Импортируем модуль для работы с потоками
import threading
```

```python
# # Объявим переменную счетчик и функцию увеличения
# counter = 0

# def increment(name):
#   global counter
#   for __ in range(100):
#     counter += 1
#     print(name, counter)

# # А также функцию уменьщения
# def disincrement(name):
#   global counter
#   for __ in range(100):
#     counter -= 1
#     print(name, counter)
```

```python
# Объявим переменную счетчик и функцию увеличения
counter = 0
# зададим блокиратор
lock = threading.Lock()

def increment(name):
  global counter
  # Установим блок
  lock.acquire()
  for __ in range(10):
    counter += 1
    print(name, counter)
  # Снимем блок
  lock.release()

# А также функцию уменьщения
def disincrement(name):
  global counter
  lock.acquire()
  for __ in range(10):
    counter -= 1
    print(name, counter)
  lock.release()
```

```python
# Создадим четыре потока
thread_1 = threading.Thread(target=increment, args=('thread_1',))
thread_2 = threading.Thread(target=disincrement, args=('thread_2',))
thread_3 = threading.Thread(target=increment, args=('thread_3',))
thread_4 = threading.Thread(target=disincrement, args=('thread_4',))
```

```python
# Запустим потоки
thread_1.start()
thread_3.start()
thread_2.start()
thread_4.start()
```

```python
# более короткая запись блокировки
def increment_1(name):
  global counter
  # Установим блок
  with lock:
    for __ in range(10):
      counter += 1
      print(name, counter)


# А также функцию уменьщения
def disincrement_1(name):
  global counter
  with lock:
    for __ in range(10):
      counter -= 1
      print(name, counter)
```

```python
# Создадим четыре потока
thread_1 = threading.Thread(target=increment_1, args=('thread_1',))
thread_2 = threading.Thread(target=disincrement_1, args=('thread_2',))
thread_3 = threading.Thread(target=increment_1, args=('thread_3',))
thread_4 = threading.Thread(target=disincrement_1, args=('thread_4',))
```

```python
# Запустим потоки
thread_1.start()
thread_3.start()
thread_2.start()
thread_4.start()
```

```python
# Использование сцепки try-finally
def increment_2(name):
  global counter
  # Установим блок
  try:
    lock.acquire()
    for __ in range(10):
      counter += 1
      print(name, counter)
  except Exception:
      pass
  finally:
    lock.release()


# А также функцию уменьщения
def disincrement_2(name):
  global counter
  try:
    lock.acquire()
    for __ in range(10):
      counter -= 1
      print(name, counter)
  except Exception:
      pass
  finally:
    lock.release()
```

```python
# Создадим четыре потока
thread_1 = threading.Thread(target=increment_2, args=('thread_1',))
thread_2 = threading.Thread(target=disincrement_2, args=('thread_2',))
thread_3 = threading.Thread(target=increment_2, args=('thread_3',))
thread_4 = threading.Thread(target=disincrement_2, args=('thread_4',))
```

```python
# Запустим потоки
thread_1.start()
thread_3.start()
thread_2.start()
thread_4.start()
```
